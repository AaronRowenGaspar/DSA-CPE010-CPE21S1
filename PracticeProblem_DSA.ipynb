ALGORITHM:

1. START
2. SUM = 0
3. WHILE THERE IS A VALUE
4. INPUT VALUE
5. IF VALUE IS EVEN
		6. SUM += VALUE
7. ELSE
		8. GO TO STEP 3
9. OUTPUT SUM
10. STOP


flowchart.png

```python
sum = 0
y = [26, 49, 98, 87, 62, 75]
print (y)
print ("---------------------------------------")
for i in range(len(y)):
  if y[i] % 2 == 0:
    sum += y[i]
print("The sum of even numbers is -->", sum)
```
